## Task 2 - Cleaning:
1. Split state and country 
2. Change timestamp datatype to date



```python
def my_country(text):
    result=text.split(',')[0]
    return result




def my_state(text):
    result=text.split(',')[0]
    return result
```

```python
log_df['state']= log_df['serverregion'].apply(my_state)
log_df['country']= log_df['serverregion'].apply(my_country)
```
